## Import packages and data

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# loading dataset 
df = sns.load_dataset("iris") 
```

## A simple Seaborn Plot
* Nice plot with one-line code
* Does the stastics such as confidence interval here
* Integrates well with matplotlib

```python
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=df)
#sns.lineplot(x="sepal_length", y="sepal_width", data=df, ci=None)
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
```

* Customize the background and style
* Utilize subplots from matplotlib

```python
# setting the color palette
sns.set_palette('vlag')
# Set the theme to dark
sns.set_style("white")
plt.subplot(211)
sns.lineplot(x="sepal_length", y="sepal_width", data=df)
  
# setting another color palette
sns.set_palette('Accent')
# Set the theme to dark
sns.set_style("dark")
plt.subplot(212)
sns.lineplot(x="sepal_length", y="sepal_width", data=df)

 
```


Count and display the categorical variables - Useful for checking data balance


```python
# count plot on single categorical variable
sns.countplot(x ='species', data = df)
```


Vertical and Horizontal Bar Plots


```python
sns.barplot(x ='species', y='sepal_length', data = df)
```

```python
sns.barplot(y ='species', x='sepal_length', data = df)
```

```python
plt.subplot(1,2,1)
sns.barplot(x ='species', y='sepal_length', data = df)
plt.subplot(1,2,2)
sns.barplot(y ='species', x='sepal_length', data = df)
# set the spacing between subplots
plt.subplots_adjust(left=0.1, bottom=0.1, right=1.2, top=0.9, wspace=0.5, hspace=0.4)
plt.show()
```

### Box Plot 
![alt text](https://github.com/dmbala/python-data-viz/blob/main/Figs/boxplot_tds.png?raw=true)

```python
sns.boxplot(x ='species', y = 'sepal_width',  data = df)
```

### Box, Violin, Strip, and Swarm Plots

```python
plt.subplot(2,2,1)
sns.boxplot(x ='species', y = 'sepal_width',  data = df)
plt.subplot(2,2,2)
sns.violinplot(x ='species', y = 'sepal_width',  data = df)
plt.subplot(2,2,3)
sns.stripplot(x ='species', y = 'sepal_width',  data = df)
plt.subplot(2,2,4)
sns.swarmplot(x ='species', y = 'sepal_width',  data = df)
plt.subplots_adjust(left=0.1, bottom=0.1, right=1.2, top=0.9, wspace=0.5, hspace=0.4)
plt.show()
```

### Histogram and Grouped Histogram

```python
plt.subplot(1,2,1)
sns.histplot(x ='sepal_length', data = df, kde='True')
plt.subplot(1,2,2)
sns.histplot(x ='sepal_length', data = df, hue='species', kde='True')
plt.subplots_adjust(left=0.1, bottom=0.1, right=1.5, top=0.9, wspace=0.4, hspace=0.1)
plt.show()
```

### Contour and Jointplot

```python
sns.kdeplot(df.sepal_length, df.sepal_width)
```

```python
sns.jointplot(df.sepal_length, df.sepal_width, kind="kde", size=7)
```

Correlation Among Variables

```python
plt.figure(figsize=(7,4))
sns.heatmap(df.corr(),annot=True,cmap='winter')
```

## Seaborn Pairgrid and Pairplot
Pairgrid: Grid of small subplots. After pairgrid, use the pairgrid.map function to draw the plots. 

Pairplot: Quick way to plot on the grid.


**Pairplot**

```python
sns.pairplot(df) # default: scatter plots for the off-diagonal and histogram distribution for the diagonal
```


**Pairplot - group by the species"


```python
sns.pairplot(df, hue='species') # adding hue changes the diagonal to kde
```

```python
sns.pairplot(df, hue='species', corner='True')
```


Histogram + KDE + Resolve by Type


```python
plt.style.use('classic')
g = sns.PairGrid(df, hue="species")
g.map_diag(sns.histplot, kde="True")
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

## FacetGrid - Looking at the subset of the data

```python

g = sns.FacetGrid(df, col="species", hue="species")
g.map(sns.scatterplot, "sepal_length", "sepal_width")
g.add_legend()
```

```python
g = sns.FacetGrid(df, col="petal_width", hue="species")
g.map(sns.scatterplot, "sepal_length", "sepal_width")
g.add_legend()
```
