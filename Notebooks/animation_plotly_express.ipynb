# GraphAnimations with Ploty
* Captures evolution of data with respect to a parameter like time


```python

import plotly.express as px

gapminder = px.data.gapminder()
gapminder.head(5)
```

```python
fig = px.scatter(gapminder, x="gdpPercap", y="lifeExp", animation_frame="year")
fig.show()
```

## Bubble Chart

```python
fig = px.scatter(gapminder, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="country", hover_name="country", 
           log_x = True, 
           size_max=45, range_x=[100,100000], range_y=[25,90])
fig.show()
```

## Choropleth - Data distributed on geographic map

```python
fig = px.choropleth(gapminder,
                    locations ="iso_alpha",
                    color ="lifeExp",
                    hover_name ="country", 
                    color_continuous_scale = px.colors.sequential.Plasma,
                    scope ="world",
                    animation_frame ="year")
fig.show()
```

```python
fig = px.bar(gapminder, 
             x ="continent", 
             y ="pop",
             color ='lifeExp',
             animation_frame ='year',
             hover_name ='country', 
             range_y =[0, 4000000000])
fig.show()
```

## Bubble chart by group

```python
fig = px.scatter(
    gapminder, 
    x ="gdpPercap", 
    y ="lifeExp", 
    animation_frame ="year", 
    animation_group ="country",
    size ="pop", 
    color ="continent", 
    hover_name ="country", 
    facet_col ="continent",
    size_max = 45,
    range_y =[25, 90],
    range_x =[-9000,50000]
)
fig.show()
```

```python
fig = px.density_contour(gapminder,
						x ="gdpPercap",
						y ="lifeExp",
						color ="continent",
						marginal_y ="histogram",
						animation_frame ='year',
						animation_group ='country',
						range_x =[-9000,50000],
						range_y =[25, 100])
fig.show()
```
