## Pandas

* pandas series - one dimentaion array that can hold any data type
* pandas dataframe - two dimensional labeled data structure that can hold columns of any data type. The data in a column is homogenous data type. 

#### Pandas Series

```python
import pandas as pd
S1 = pd.Series([1, 2, 3, 4, 5])
S2 = pd.Series([1, 2, 3, 4, 5])
S3 = S1 + S2 # Vector operations work
print(S3)
print(S3.index) 
print(S3.values)
print("S[2]= ", S3[2]) # Access by Index
```

```python
fruits1 = ['apples', 'oranges', 'cherries', 'pears']
fruits2 = ['peaches', 'oranges', 'cherries', 'apples']

S1 = pd.Series([1, 2, 3, 4], index=fruits1) # Labels for the index
S2 = pd.Series([4, 2, 3, 1], index=fruits2) 
print(S1)
print(S2)
print("S1[apples]=", S1['apples']) # Access by labeled index
print("S1.iloc[0]", S1.iloc[0])
```

```python
S3 = S1 + S2
print(S3) # Vector operations are based on the labeled index
```

```python
print(S3.dropna())
```

## Object Oriented vs Pyplot/Functional Interface
* OO Interface offers fine grained control
* Pyplot offers a familiar to MATLAB and easy interface

**OO Interface**

```python
fig = plt.figure()  # an empty figure with no Axes
```

```python
fig, ax = plt.subplots()  # a figure with a single Axes
```

```python
x = np.arange(-20., 20., 0.1)
y = np.sin(x)/x

fig, ax = plt.subplots() # Creates instances - fig, ax
ax.plot(x,y) # change the ax instance
ax.set_title("Oscillating function")
plt.show()
```

**PyPlot/Functional Interface**

```python
x = np.arange(-20., 20., 0.1)
y = np.sin(x)/x

plt.plot(x,y) #Pyplot Interface
plt.title("Oscillating Function")
```

Multiple Curves

```python
x = np.arange(-20., 20., 0.1)
y1 = np.sin(x)/x
y2 = np.sin(2.0*x)/x

plt.plot(x, y1, x, y2)
```

```python
x = np.arange(-20., 20., 0.1)
y1 = np.sin(x)/x
y2 = np.sin(2.0*x)/x

plt.plot(x, y1)
plt.plot(x, y2)
```

Differentiate the curves by colors, symbols, and legends

```python
x = np.arange(-20., 20., 0.1)
y1 = np.sin(x)/x
y2 = np.sin(2.0*x)/x

plt.plot(x,y1, 'r-v',x,y2, 'b-o') # red-triangle, blue-dot
plt.legend(["slow oscillation", "fast oscillation"]) #Legends
```

Annotate the plot and curves - Title, labels, legends, colors, symbol, lines, ...

```python

x = np.arange(-20., 20., 0.1)
y1 = np.sin(x)/x
y2 = np.sin(2.0*x)/x

plt.plot(x,y1, 'r-v',x,y2, 'b-o')                    # red-triangle, blue-dot
plt.legend(["slow oscillation", "fast oscillation"]) #Legends

plt.title ("Oscilating function")
plt.xlabel('radians')               # x label
plt.ylabel('amplitude')                              # y label
plt.axhline(y=0, color='k')
plt.grid(False)
```

### Grouped Bar Charts

```python
last_week = [22, 35, 30, 34, 27, 30]
this_week = [25, 31, 36, 24, 25, 33]
names = ['Mary', 'Larry', 'Tory', 'Barry', 'Bill', 'Will']


width = 0.25   
ticks = np.arange(len(names))    
plt.bar(ticks, last_week, width, label='Last week')
plt.bar(ticks + width, this_week, width, align="center", label='This week')
plt.xticks(ticks + width/2,names, fontsize=20, rotation=45) # Adjust xtics to the center and rotate by 45 degrees
plt.legend(["Last Week", "Current Week"])
plt.ylabel("Number of Points", fontsize=20)
plt.show()
```

```python
rows, cols = 2, 4
fig, ax = plt.subplots(rows, cols, sharex='col', sharey='row')
                    
for row in range(rows):
    for col in range(cols):
        ax[row, col].text(0.5, 0.5, str((row,col)),fontsize=22,ha='center')
                      
plt.show()
```

```python

fig_index = 1            
for row in range(0, rows):
    for col in range(0, cols):
         plt.subplot(rows,cols,fig_index)
        
         plt.text(0.5,0.5, str((row,col,fig_index)), ha='center', fontsize='16', color='green')   
         if row == 2: 
             plt.xticks([])
         if col > 0:
             plt.yticks([])      
         fig_index = fig_index + 1            
plt.show()
```

### Subplots can be of different types - scaterplot, bar, etc.

```python
x = np.random.normal(0, 1, 1000)
y = x**2.0
y1 = x**3.0

plt.subplot(2,2,1)
plt.scatter(x,y)

plt.subplot(2,2,2)
plt.hist(y, color='black')

plt.subplot(2,2,3)
plt.scatter(x,y1, color='green')

plt.subplot(2,2,4)
plt.hist(y1, color='red')
```

Each subplot can be independtly adjusted

```python
x = np.random.normal(0, 1, 1000)
y = x**2.0
y1 = x**3.0

plt.subplot(2,2,1)
plt.scatter(x,y)
plt.xticks([])  # Removes the xticks

plt.subplot(2,2,2)
plt.hist(y, color='black')
# fine grained control on ticks
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off

plt.subplot(2,2,3)
plt.scatter(x,y1, color='green')

plt.subplot(2,2,4)
plt.hist(y1, color='red')
```

### Contour Plots

```python
fig = plt.figure(figsize=(6,5))
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height]) 

start, stop, n_values = -8, 8, 800

x_vals = np.linspace(start, stop, n_values)
y_vals = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x_vals, y_vals)


Z = np.sqrt(X**2 + Y**2)

cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)

ax.set_title('Contour Plot')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```

3D Plots with Contours

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)
cset = ax.contour(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)

ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100)

plt.show()
```

## Plots from Panda Dataframes
* Good representation of tabular data that have headings for each column. 
* Inhomogenous column types - common with Excel, SQL table, ...
* Can arbitray switch between index and label to access the data
* Easy to filter data
* Fast compared to list of lists


```python
## Example from SW material 
import pandas as pd
#data = pd.read_csv('https://raw.githubusercontent.com/ay7n/RUCIL-PY-JUPYTER-DATA-VISUALIZATION/master/Data/gapminder_gdp_oceania.csv', index_col='country')
data = pd.read_csv('https://raw.githubusercontent.com/dmbala/python-data-viz/master/Data/gapminder_gdp_oceania.csv', index_col='country')
# Extract year from last 4 characters of each column name
# The current column names are structured as 'gdpPercap_(year)', 
# so we want to keep the (year) part only for clarity when plotting GDP vs. years
# To do this we use strip(), which removes from the string the characters stated in the argument
# This method works on strings, so we call str before strip()

years = data.columns.str.strip('gdpPercap_')
# Convert year values to integers, saving results back to dataframe
data.columns = years.astype(int)
```

* By default, DataFrame.plot plots with the rows as the X axis.
* We can transpose the data in order to plot multiple series.

```python
data.T
```

```python
plt.style.use('ggplot')
data.T.plot(kind='bar')
plt.ylabel('GDP per capita')
```
